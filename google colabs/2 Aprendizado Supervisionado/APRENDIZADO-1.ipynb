# Medindo a performance do modelo

- Poderiamos utilizar os dados usados para ajustar/treinar o modelo
    - Não seria um indicativo de abilidade de generalizar
- Dividiremos os dados em conjunto de Treino e conjunto de Teste
    - Fit/Treinar o classificador no conjunto de dados de treino
    - Realizar predições no conjunto de dados de teste
    - Comparar predições com os rótulos conhecidos

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=21, stratify=y)
```

- random_state é a seed para reproduzir a mesma geração randomica
- o padrão do teste_size é 75(treino) e 25(teste), aqui está setado para 30% 
- stratify faz com que o train set e o teste set tenham suas distribuições da mesma forma que nos dados completos por isso o y (lista ou array contendo os rótulos do dados completos) 

```python
knn = KNeighborsClassifier(n_neighbors=8)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
print("Test set predictions:\n {}".format(y_pred))
```

```python
knn.score(X_test, y_test)
```

# Complexidade do Modelo
- k alto = margem de decisão mais suave = modelo de menor complexidade
- k baixo = modelo mais complexo = pode levar ao overfitting (Sensibilidade do modelo a ruídos (noise) ou a valores específicos dos dados utilizados, que não refletem a tendência geral dos dados)


# Model Complexity Curve
- Se o valor de k for aumentado ainda mais, o modelo vai ficando ainda mais simples... levando a uma pior performance tanto em treino quanto em teste... conhecido como Underfitting.

```python
# Setup arrays to store train and test accuracies
neighbors = np.arange(1, 9)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))
```

```python
# Loop over different values of k
for i, k in enumerate(neighbors):
    # Setup a k-NN Classifier with k neighbors: knn
    knn = KNeighborsClassifier(n_neighbors=k)

    # Fit the classifier to the training data
    knn.fit(X_train, y_train)
    
    #Compute accuracy on the training set
    train_accuracy[i] = knn.score(X_train, y_train)

    #Compute accuracy on the testing set
    test_accuracy[i] = knn.score(X_test, y_test)

# Generate plot
plt.title('k-NN: Varying Number of Neighbors')
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```
